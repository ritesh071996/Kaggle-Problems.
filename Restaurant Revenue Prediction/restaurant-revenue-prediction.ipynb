```python
import time
from datetime import datetime as dt
# train
all_diff = []
for date in train["Open Date"]:
    diff = dt.now() - dt.strptime(date, "%m/%d/%Y")
    all_diff.append(int(diff.days/1000))

train['Days_from_open'] = pd.Series(all_diff)
print(train.head())
```

```python
import time
from datetime import datetime as dt
# train
all_diff = []
for date in test["Open Date"]:
    diff = dt.now() - dt.strptime(date, "%m/%d/%Y")
    all_diff.append(int(diff.days/1000))

test['Days_from_open'] = pd.Series(all_diff)
print(test.head())
```

```python
y_train = train["revenue"]
X_train = train.drop(["Id","Open Date","City","revenue"],axis = 1)
X_test = test.drop(["Id","Open Date","City"],axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_clf = LogisticRegression(random_state=42,class_weight='balanced')
log_clf.fit(X_train,y_train)
```

```python
pred_log=log_clf.predict(X_test)
pred_log
```

```python
submission1 = pd.read_csv("sampleSubmission.csv")
submission1["Prediction"] = pred_log
submission1.to_csv("submissionlog.csv",index = False)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train,y_train)
```

```python
pred_tree = tree_model.predict(X_test)
pred_tree
```

```python
submission2 = pd.read_csv("sampleSubmission.csv")
submission2["Prediction"] = pred_tree
submission2.to_csv("submissiontree.csv",index = False)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier(n_estimators=500,random_state = 1)
forest_model.fit(X_train,y_train)
```

```python
pred_forest = forest_model.predict(X_test)
pred_forest
```

```python
submission3 = pd.read_csv("sampleSubmission.csv")
submission3["Prediction"] = pred_forest
submission3.to_csv("submissionforest.csv",index = False)
```
